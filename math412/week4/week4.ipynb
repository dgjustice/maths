```python
G = nx.Graph()
G.add_nodes_from('abcde')
G.add_edges_from([
    ('a', 'b'),
    ('a', 'e'),
    ('b', 'e'),
    ('b', 'c'),
    ('c', 'd'),
    ('d', 'e'),
])
nx.draw(G, with_labels=True)
```

```python
nx.to_numpy_array(G)
```

```python
C = nx.laplacian_matrix(G).toarray()
print(C)
m = sympy.Matrix(C)
m.cofactor(3, 1)
```
