```python
import networkx as nx
from networkx.algorithms import bipartite
from IPython.display import display

from sympy import *
import numpy as np

# x, y, z = symbols('x y z')
init_printing(use_unicode=True)
```

```python
G = nx.Graph()
G.add_nodes_from('abcdefgh')
elist = [
    ('a', 'e'),
    ('b', 'e'),
    ('c', 'e'),
    ('d', 'e'),
    ('a', 'f'),
    ('b', 'f'),
    ('c', 'f'),
    ('d', 'f'),
    ('g', 'e'),
    ('h', 'e'),
    ('g', 'f'),
    ('h', 'f'),
    ('a', 'b'),
    ('c', 'd'),
    ('g', 'h'),
]
G.add_edges_from(elist)
nx.draw(G, with_labels=True)
```

```python
adj = nx.adjacency_matrix(G).todense()
print(adj)

adj2 = np.delete(adj, np.s_[4:6], 0)
adj2 = np.delete(adj2, np.s_[4:6], 1)
print(adj2)
```

```python
D = nx.DiGraph()
D.add_nodes_from('abcdefg')
elist = [
    ('a', 'f'),
    ('f', 'g'),
    ('g', 'e'),
    ('e', 'f'),
    ('e', 'd'),
    ('d', 'b'),
    ('b', 'a'),
    ('d', 'c'),
    ('c', 'b'),
]
D.add_edges_from(elist)
nx.draw(D, pos=nx.kamada_kawai_layout(D), with_labels=True)
```

```python
print(D.nodes)
nx.adjacency_matrix(D).todense()
```
